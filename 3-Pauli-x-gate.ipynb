```python
from qiskit import *
from qiskit.tools.visualization import array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

### Creating a quantum circuit with one qubit and applying Pauli-X gate

```python
circuit = QuantumCircuit(1)
# circuit.ry(pi/4, 0)
# circuit.x(0)
circuit.x(0)
print(circuit)
circuit.draw(output='mpl')
```

### Creating a quantum circuit with two qubits and applying rotation and Pauli-Z gate

```python
circuit = QuantumCircuit(2)
circuit.ry(pi/4, [0,1])
# circuit.y(1);

circuit.z(0)
circuit.draw(output='mpl')
```

### Simulating the circuit using statevector_simulator and displaying the statevector

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
statevector = result.get_statevector()
array_to_latex(statevector, prefix="\\text {statevector =}\n")
```

### Plotting the Bloch multivector

```python
plot_bloch_multivector(statevector)
```
