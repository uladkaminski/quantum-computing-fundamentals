```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
from math import pi
%matplotlib inline
```

### Creating a quantum circuit with one qubit and applying Hadamard gate

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
circuit.h(0)
# circuit.h(0)
# circuit.y(0)
# circuit.h(0)
# circuit.x(0)
circuit.draw(output='mpl')
```

### Simulating the circuit using statevector_simulator and displaying the statevector

```python
simulator = Aer.get_backend("statevector_simulator")
statevector = execute(circuit, backend=simulator).result().get_statevector()
array_to_latex(statevector)
```

### Plotting the Bloch multivector

```python
plot_bloch_multivector(statevector)
```

### Measuring the qubit

```python
circuit.measure(0, 0)
circuit.draw(output='mpl')
```

### Simulating the circuit using qasm_simulator and plotting the histogram

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots= 1024).result()
plot_histogram(result.get_counts())
```
