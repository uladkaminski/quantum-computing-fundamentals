```python
from qiskit import *
from qiskit.tools.visualization import array_to_latex, plot_bloch_multivector
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.h([0, 1, 2])
# circuit.s(1)
# circuit.sdg(2)
circuit.t(1)
circuit.tdg(2)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend("statevector_simulator");
statevector = execute(circuit, backend=simulator).result().get_statevector()
plot_bloch_multivector(statevector)
```

# Parameterized Rotation Gates

```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
```

```python
circuit = QuantumCircuit(3)
circuit.rx(2, 0)
circuit.rx(pi/3, 1)
circuit.h(2)
# circuit.rz(pi/3, 2)
circuit.draw(output='mpl')
```

```python
backend = Aer.get_backend("statevector_simulator")
statevector = execute(circuit, backend=backend).result().get_statevector()
```

```python
plot_bloch_multivector(statevector)
```

```python
# Challenge
from qiskit.tools.visualization import plot_histogram
```

```python
circuit = QuantumCircuit(3) 

circuit.y(0)
circuit.h(1)
circuit.ry(pi/3, 2)


circuit.measure_all()
circuit.draw(output='mpl')
```

```python
backend = Aer.get_backend("qasm_simulator")
result = execute(circuit, backend=backend, shots=1000).result()
plot_histogram(result.get_counts())
```
