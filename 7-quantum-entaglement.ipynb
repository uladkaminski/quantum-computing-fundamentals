```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
```

### Bell States

```python
circuit = QuantumCircuit(2)

# [1/sqrt(2), 0, 0 , 1/sqrt(2)]
# circuit.h(0)
# circuit.cnot(0, 1)

# [0, 1/sqrt(2),1/sqrt(2), 0]
# circuit.x(1)
# circuit.h(0)
# circuit.cnot(0, 1)

# [1/sqrt(2), 0, 0 , -1/sqrt(2)]
# circuit.x(0)
# circuit.h(0)
# circuit.cnot(0, 1)

# -[0, 1/sqrt(2), -1/sqrt(2), 0]
circuit.x([0,1])
circuit.h(0)
circuit.cnot(0, 1)


circuit.draw(output='mpl')
```

```python
backend = Aer.get_backend("statevector_simulator")
result = execute(circuit, backend=backend).result()
statevector = result.get_statevector()
array_to_latex(statevector)
```

```python
circuit.measure_all()
backend = Aer.get_backend("qasm_simulator")
result = execute(circuit, backend=backend, shots=1024).result()
plot_histogram(result.get_counts())
```

### Challenge (GHZ state)



```python
circle = QuantumCircuit(3)
circle.h(0)
circle.cx(0, 1)
circle.ccx(0, 1, 2)
circle.draw(output='mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
statevector = execute(circle, backend=backend).result().get_statevector()
array_to_latex(statevector)
```

```python
circle.measure_all()

backend = Aer.get_backend('qasm_simulator')
result = execute(circle, backend=backend, shots=1024).result()

plot_histogram(result.get_counts())
```
